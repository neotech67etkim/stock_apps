```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid_f(x):
    return 1 / (1+np.e**(-x))

def tanh_f(x):
    return np.tanh(x)

def relu_f(x):
    return np.where(x<=0, 0, x)

def softplus_f(x):
    return np.log(1+np.exp(x))
```

```python
x = np.linspace(-5,5)
print(x)
```

```python
plt.title('sigmoid Function')
sigmoid_y = sigmoid_f(x)
plt.plot(x,sigmoid_y)
plt.show
```

```python
plt.title('tanh Function')
tanh_y = tanh_f(x)
plt.plot(x,tanh_y)
plt.show
```

```python
plt.title('relu Function')
relu_y = relu_f(x)
plt.plot(x,relu_y)
plt.show
```

```python
plt.title('softplus Function')
softplus_y = softplus_f(x)
plt.plot(x,softplus_y)
plt.show
```
